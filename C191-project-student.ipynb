```python
#Covide Data 
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
df
```

```python
df = pd.DataFrame(dataframe, columns =['total_cases','country'])
df.replace(" ", float("NaN"), inplace=True)

df = df.dropna()

sorted_df = df.sort_values(by=['total_cases'])
sorted_df

```

```python
least_cases = df.tail(5)
print(least_cases)

name = least_cases['country']
cases = least_cases['total_cases']

plt.xlabel("Country name")
plt.xticks(rotation='vertical')
plt.ylabel("No. of cases")

label = name
value = cases
plt.bar(label,value,width=0.4,color=('red', 'blue', 'green', 'pink', 'black', 'yellow'))
```

```python
df = pd.DataFrame(dataframe, columns =['total_deaths','country'])
df.replace(" ", float("NaN"), inplace=True)

df = df.dropna()

sorted_df = df.sort_values(by=['total_deaths'])
sorted_df

```

```python
most_deaths = df.head(5)
print(most_deaths)

name = most_deaths['country']
deaths = most_deaths['total_deaths']

plt.xlabel("Country name")
plt.xticks(rotation='vertical')
plt.ylabel("No. of deaths")

label = name
value = deaths
plt.bar(label,value,width=0.4,color=('red', 'blue', 'green', 'pink', 'black', 'yellow'))
```
